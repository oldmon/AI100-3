# Rectified Linear Unit- Relu 

f(x)=max(0,x)


```python
import numpy as np
from numpy import *
import matplotlib.pylab as plt
%matplotlib inline

def relu(x):
    return np.where(x<0,0,x)

#Sigmoid 微分
def drelu(x):
    return np.where(x<0,0,1)
    
x = plt.linspace(-4,4,100)
plt.plot(x, relu(x), 'b', label='linspace(-4,4,10)')
plt.plot(x, drelu(x), 'g', label='linspace(-4,4,10)')
plt.grid()
plt.title('ReLU Function')
plt.text(-2, 3, r'$ReLU(x)=max(0,x)$', fontsize=15)
plt.gca().xaxis.set_major_locator(plt.MultipleLocator(1))
plt.gca().yaxis.set_major_locator(plt.MultipleLocator(1))
plt.show()
```
